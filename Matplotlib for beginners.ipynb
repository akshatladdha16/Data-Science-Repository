# Python Visualization Libraries

Matplotlib,seaborn are the most used visualisation libraies we use to visualise our data to gain better insights which we cannot see just by looking at data

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('darkgrid')
import numpy as np
import pandas as pd
```

```python
x=np.arange(0,10)
y=np.arange(50,60)
```

```python
# scatter plot
plt.scatter(x,y,c='r')
plt.xlabel('x Axis')
plt.ylabel('y Axis')
plt.title('Realtion btw x and y')

# to save this plot
plt.savefig('Test.png')
```

```python
# plot
y=x*x
plt.plot(x,y,'g*',linestyle='dashed',markersize=10,linewidth=2)
```

```python
# creating susbplots: multiple plots in one frame
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'b*--')
```

```python
print(np.pi)
x=np.arange(0,4*np.pi,0.001)
y=np.sin(x)
plt.title('sine waveform')
plt.plot(x,y,'b')
```

```python
# bar plots in matplotlib(still seaborn bar plots are uch better than this and we will see them later in this notebook)
x=[2,8,10]
y=[11,16,9]
x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color='g')
```

```python
# Histograms: used to compare density on y axis
a=np.array([22,87,5,43,56,73,55,54,11,28,24,79,32])
plt.hist(a,bins=20)
# bins specify the density of histogram which shows what's the count in that particular range
```

```python
# Box plot: to find out percentile
data=[np.random.normal(0,std,100) for std in range(1,4)]
# range 1,4 so 3 values will get listed here in boxplot
plt.boxplot(data,vert=True,patch_artist=True)
```

these circles are outliers 

```python
# Pie Chart
# plotting data
labels='Python','C++','Ruby','Java'
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0.1,0.1,0.1)# here exploding 1st slice only means python one
plt.pie(sizes,explode=explode,labels=labels,colors=colors,
       autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.title('Coding languages poularity')
```
