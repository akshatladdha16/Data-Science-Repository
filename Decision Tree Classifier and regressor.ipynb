```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris.data
```

```python
iris.target
```

```python
# or we can use seaborn which gives us in form of dataframe
df=sns.load_dataset('iris')
df.head()
```

```python
x=df.iloc[:,:-1]
y=iris.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
# we can see 4 indeendent features here
from sklearn.tree import DecisionTreeClassifier
# always check sklearn website for parameters to set in our classifier
#as we are going with none for hyperparameters it is post pruning technique
classifier=DecisionTreeClassifier()
classifier.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(classifier,filled=True) #filled gives boxes with info given
```

now in this case on left split we get 31,0,0 so we don't need any further split as we reached pure split

we can perform post pruning on this which can avoid unnecessary splits. we can use hyperparameters like max_depth 

```python
from sklearn.tree import DecisionTreeClassifier
# always check sklearn website for parameters to set in our classifier
#as we are going with none for hyperparameters it is post pruning technique
classifier_pp=DecisionTreeClassifier(max_depth=2)
classifier_pp.fit(x_train,y_train)
```

```python
#predictions
y_pred_pp=classifier_pp.predict(x_test)
y_pred_pp
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(classifier_pp,filled=True) 
```

```python
from sklearn.metrics import accuracy_score,classification_report
score=accuracy_score(y_pred_pp,y_test)
print(score)
```

```python
print(classification_report(y_pred_pp,y_test))#after doing pruning
```
