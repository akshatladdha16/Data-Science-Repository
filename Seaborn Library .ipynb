```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_style('darkgrid')
import warnings
warnings.filterwarnings('ignore')
```

jointplot used to do bivariate analysis
and if more than 3 features are there we use pairplot to get insights btw many features

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.info()
# small dataset on restaurants having info about these things
# here tip is our dependent feature , other are independent
```

# Correlation with headtmap

```python
# used to find inter relationship btw all features.
# only possible when values are int or float , not applicable to objects
df.corr()
# we will get 3 info as only 3 columns has integer or float data 
```

here it shows that if total bill is increasing tip will also increase as it it 67% correlated

```python
sns.heatmap(df.corr())
```

this is very important when we do feature selection, eg. when we there are 2 indep features and corr of those two with dpenedent feature is same then we can drop onw feature to optimise the algo

```python
# Joinplot: used for univariate or bivariate analysis
sns.jointplot(x='tip',y='total_bill',data=df,color='blue',kind='reg')
# regression line is best fit line and side hist lines are probablity density analysis
```

```python
# now more than two features analysis
# Pairplot: do PnC of all the features
sns.pairplot(df)
```

```python
# now if we want to do analysis of all features with different feature and classify the points here
# lets say sex feature analysis on other features
sns.pairplot(df,hue='sex')
# noiw all points are showing classification on the basis sex feature
```

```python
# Distplot: used to check the distribution of the column features
sns.distplot(df['tip'],kde=False)
# kde is kernel density estimation if we make is True it will 
# show the densityy curve instead of count
```

```python
# Now we will look for visualisation for categoricl features
# boxplot, countplot,violonplot
# countplot
sns.countplot('day',data=df)
```

```python
# barplot : same as countplot but here we are having both xa nd y values
sns.barplot(x='sex',y='total_bill',data=df)
# another univariate analysis but catgorical feature
# boxplot gives u percentile distribution
```

```python
# boxplot
sns.boxplot(x='day',y='total_bill',data=df)
```

```python
sns.boxplot(x='sex',y='total_bill',data=df,hue='smoker')
# using hue so we are having different plots for sex features parameteres
```

# violin plot
