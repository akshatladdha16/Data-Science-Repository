open source library for data analysis and we can use other libraries of python with pandas. Pandas uses power and speed of numpy(numpy is written in c++). can handle big data very easily

two types of data structures in pandas:Series(1dimensional array with indexes, stores single colu or row of data in data frame), data frame(tabular spreadsheet which contain one or multiple columns)

```python
newdf=pd.DataFrame(np.random.rand(334,5),index=np.arange(334))
print(type(newdf))
newdf.head()
```

```python
newdf.describe()
```

```python
newdf[0][0]='akshat'
# we are changing datatype in dataframe so pandas change it for us in df
print(newdf.dtypes)
# getting object as our first datatype
newdf.head()
```

```python
newdf.columns
```

```python
# to convert this in numpy array
num=newdf.to_numpy()
num
```

```python
newdf.head()
```

```python
newdf.sort_index(axis=0,ascending=False).head()
# sorting in descending order
```

```python
newdf2=newdf 
# here newdf2 is just a view or pointer to newdf
# there's no other dataframe named newdf2, so any changes we do in newdf2 will be there in newdf
newdf2[0][0]=973
newdf.head()
# see change happend in newdf when we altered newdf2 
```

```python
# to remove this possibility we copy data frame
newdf2=newdf.copy()
newdf2[0][0]=2345
newdf.head()
# see now there is no change in newdf after altering data in newdf2
```

```python
# to set any value in data frame we should avoid situation of view(copy) of df, so we use loc function

newdf.loc[0,'A']=654
newdf.head()
```

```python
# to drop last column; axis=1 column and axis=0 is row
newdf=newdf.drop(0,axis=1)
newdf.head()
```

```python
# to change column names
newdf.columns=list("ABCDE")
newdf.head()
```

```python
# more loc applications 
newdf.loc[[1,2],['C','D']]
# to access any particular row(first) and column of dataframe
```

```python
# to get all the colums use :
newdf.loc[[1,2],:]
```

```python
# to get all rows
newdf.loc[:,['A']]
```

```python
# filtering data using loc
newdf.loc[(newdf['A']<0.3)&(newdf['C']>0.1)]
```

```python
newdf.iloc[0,4]
# here we can access a particular data member without particulaly using name of rows and columns
```

```python
newdf.iloc[[0,5],[1,2]]
```

```python
# dropping columns and rows from dataset
print(newdf.head()) 
newdf.drop(['D'],axis=1,inplace =True)
# inplace True mean save the edits in dataframe also
newdf.head()
# so now we can see our D column is dropped from our dataset
```

```python
newdf.reset_index(inplace=True,drop=True)
newdf.head()
# to reset index if you want to reset indexing after any changes
```

```python
newdf.isnull().sum()
# so see if any null values in our column or not
```

```python
# use loc to set any values coz sometimes there might be chaining or your edit won't get saved so use this method
newdf.loc[:,['B']]=None
newdf['B'].isnull().sum()
# so now we can see all rows in B column are set to null
```
